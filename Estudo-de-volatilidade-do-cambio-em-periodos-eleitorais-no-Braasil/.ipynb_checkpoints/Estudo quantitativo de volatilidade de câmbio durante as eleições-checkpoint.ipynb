## Estudo Quantitativo de Volatilidade de Câmbio durante as Eleições Presidenciais no Brasil

Acompanhar a dinâmica do mercado e compreender fatores que podem ou não prejudicar certos investimentos é fundamental para que um investidor obtenha bons resultados.

Em longo prazo, é notável uma desvalorização mais acelerada do real com relação ao dólar, bem como comportamentos atípicos resultando de acontecimentos marcantes, a exemplo da quebra da bolsa de valores. Além disso, outras variáveis possuem uma clara associação com o valor do câmbio:
   - taxa de juros no Brasil: um aumento na taxa de juros atrai os investidores, fazendo com que o real valorize
   - taxa de juros nos EUA: analogamente, um aumento nessa taxa atrai os investidores para os EUA, fazendo com que o dólar seja valorizado e, consequentemente, o valor do câmbio aumente
   - crises financeiras: neste casos, os investidores tendem a alocar suas aplicações em busca de uma volatilidade menor, o que tende a ocasionar uma alta no dólar. O Brasil é um país emergente sujeito a sofrer os efeitos de crises, sendo estas internas ou não.
   - inflação: reflete a perda do poder de compra, isto é, a desvalorização do real. Portanto, o aumento da inflação implica no aumento do valor da compra de dólar em real

```python
df_cambio = pd.read_csv('cambio_diario_bacen.csv',
                    encoding = 'utf-8',
                     sep = ';', names = ['Data', 'Câmbio']).apply(lambda x: x.str.replace(',','.'))

df_cambio.drop(index = [0, df_cambio.index[-1]], 
        axis = 0, 
        inplace = True)

df_cambio['Câmbio'] = df_cambio['Câmbio'].astype(float)

df_cambio['Data'] = pd.to_datetime(df_cambio['Data'], dayfirst = True)

df_cambio.set_index('Data', inplace = True)
```

```python
juros_selic = pd.read_csv('selic_aa.csv',
                    encoding = 'utf-8',
                     sep = ';', names = ['Data', 'Taxa Selic % a.a']).apply(lambda x: x.str.replace(',','.'))

juros_selic.drop(index = [0, juros_selic.index[-1]], 
        axis = 0, 
        inplace = True)

juros_selic['Taxa Selic % a.a'] = juros_selic['Taxa Selic % a.a'].astype(float)

juros_selic['Data'] = pd.to_datetime(juros_selic['Data'], dayfirst = True)

juros_selic.set_index('Data', inplace = True)
```

Com a corrida presidencial se aproximando, é possível que a taxa de câmbio sofra alterações em decorrência disso? 

Para responder essa pergunta, analisamos dados históricos para tentar identificar uma possível relação entre período eleitoral e taxa de câmbio.

### Comportamento anual do valor de câmbio real/dólar

O desvio padrão é uma medida que descreve uma variação em torno do valor médio. Quanto maior o desvio, maior a flutuação. Ao calcular essa medida com relação à média anual do câmbio, estaremos olhando para uma medida de variação absoluta com relação ao valor do câmbio. Assim, o desvio padrão relativo é uma estatística descritiva que visa, como o próprio nome diz, expressar uma medida relativa da variação em pontos percentuais.

```python
cambio_statistics = pd.DataFrame(index = df_cambio.index.year.unique(), columns = ['Desvio Padrão', 'Valor Máximo', 'Valor Mínimo', 'Valor Médio', 'Desvio Padrão Relativo'])
```

```python
for i in cambio_statistics.index:
    cambio_statistics['Desvio Padrão'][i] = np.std(df_cambio['Câmbio'].loc[df_cambio.index.year == i])
    cambio_statistics['Valor Máximo'][i] = np.amax(df_cambio['Câmbio'].loc[df_cambio.index.year == i])
    cambio_statistics['Valor Mínimo'][i] = np.amin(df_cambio['Câmbio'].loc[df_cambio.index.year == i])
    cambio_statistics['Valor Médio'][i] = np.mean(df_cambio['Câmbio'].loc[df_cambio.index.year== i])
    
cambio_statistics['Desvio Padrão Relativo'] = cambio_statistics['Desvio Padrão']/cambio_statistics['Valor Médio']
```

A tabela abaixo ordena os anos em ordem decrescente com relação ao desvio padrão relativo do valor de câmbio.

```python
cambio_statistics.sort_values(ascending = False, by = 'Desvio Padrão Relativo')
```

```python
#exportando tabela para o excel

file_name = 'Estatisticas_Cambio_real-dolar2.xlsx'
cambio_statistics.sort_values(ascending = False, by = 'Desvio Padrão Relativo').to_excel(file_name)
```

Nessa perspectiva, com exceção de 2002, os anos de eleição presidencial não parecem ter alta correlação com a volatilidade do câmbio.

### Variância do valor de câmbio em anos eleitorais

A variância móvel num intervalo dado durante anos eleitorais poderia fornecer uma intuição sobre a volatilidade conforme o período eleitoral se aproxima.

```python
cambio_ano_eleitoral = pd.DataFrame()

for i in range(1998, 2026, 4):
    cambio_ano = pd.DataFrame(df_cambio.loc[df_cambio.index.year == i], )
    cambio_ano.rename(columns = {'Câmbio': '{0}'.format(i),}, inplace = True)
    cambio_ano.index = cambio_ano.index.strftime('%m-%d')
    cambio_ano_eleitoral = pd.concat([cambio_ano_eleitoral, cambio_ano], axis = 1)
    
cambio_ano_eleitoral = cambio_ano_eleitoral.sort_index()
```

```python
cambio_ano_eleitoral
```

```python
ax = cambio_ano_eleitoral.plot(color = ['gray', 'c', 'r', 'b', 'orange', 'g', 'gold'])
ax.set_xlabel('Jan   Fev   Mar   Abr   Mai   Jun   Jul   Ago   Set   Out   Nov   Dez')
ax.set_ylabel('Dólar em real')
ax.set_xticklabels([])
ax.legend(bbox_to_anchor=(1.1, 1.05))

plt.suptitle('Comportamento do Câmbio em Anos Eleitorais', y = 1.05, fontsize = 18)
# plt.show()
```

 <font size = '1'> Fonte: BCB-Demab, 2022</font>

O período eleitoral por si só não demonstra correlação direta com a volatilidade, mas pode ter relação causal dependente dos resultados, medidas e expectativas resultantes da eleição. Tais fatores acabam relacionando-se diretamente com as variáveis mencionadas inicialmente.
